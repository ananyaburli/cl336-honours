```python
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt
```

```python
def odes(v,t):
    '''
    dx/dt=-ay/ ((x-1)**2 + y**2)
    dy/dt=a(x-1)/((x-1)**2 + y**2)
    
    '''
    a=5
    
    x=v[0]
    y=v[1]
    
    dxdt=-a*y/((x-1)**2 + y**2)
    dydt=a*(x-1)/((x-1)**2 + y**2)
    
    return [dxdt,dydt]
    
```

```python
def eval_xy(v0,t):

    #EVALUATE
    v= odeint(odes,v0,t)
    
    x=v[:,0]
    y=v[:,1]
    return [x,y]
```

```python
#TIME VECTOR
num_tracers=2500; #number of tracers
t=np.linspace(0,10,100)
s=t.size
X_mat=np.zeros((num_tracers,s))
Y_mat=np.zeros((num_tracers,s))

#consider the initial patch of tracer particles to be a circle centred at the origin whose radius is 0.1

for i in range(num_tracers):
    #INITIAL CONDITIONS
    v0=[0.1*np.cos((i*np.pi/180)*(360/num_tracers)),0.1*np.sin((i*np.pi/180)*(360/num_tracers))]
    [x,y]=eval_xy(v0,t)
    
    #Now we will store all X's as a matrix and all Y's as a matrix. Each row corresponds to a tracer and each column corresponds on a time point.
    X_mat[i]=x
    Y_mat[i]=y
```

```python
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt
```

```python
def odes_2(v,t):
    '''
    dx/dt=-ay/ ((x-1)**2 + y**2)
    dy/dt=a(x-1)/((x-1)**2 + y**2)
    
    '''
    a=5
    
    x=v[0]
    y=v[1]
    
    dxdt=-a*y/((x-1)**2 + y**2)
    dydt=a*(x-1)/((x-1)**2 + y**2)
    
    return [dxdt,dydt]


def odes_1(v,t):
    '''
    dx/dt=-ay/ ((x+1)**2 + y**2)
    dy/dt=a(x+1)/((x+1)**2 + y**2)
    
    '''
    a=5
    
    x=v[0]
    y=v[1]
    
    dxdt=-a*y/((x+1)**2 + y**2)
    dydt=a*(x+1)/((x+1)**2 + y**2)
    
    return [dxdt,dydt]
    
    
```

```python
def eval_xy(v0,t):

    #EVALUATE
    if (t[-1]% 0.5)==0:
        v= odeint(odes_2,v0,t)
    else:
        v= odeint(odes_1,v0,t)
    x=v[:,0]
    y=v[:,1]
    return [x,y]
```

```python
#TIME VECTOR
num_tracers=2500; #number of tracers
#t=np.linspace(0,10,200)
s=160
X_mat=np.zeros((num_tracers,s))
Y_mat=np.zeros((num_tracers,s))

#consider the initial patch of tracer particles to be a circle centred at the origin whose radius is 0.1

for j in range(40):
    t=np.linspace(0.25*j,0.25*j+0.25,4)
    for i in range(num_tracers):
        
        #INITIAL CONDITIONS
        if(j==0):
            v0=[0.1*np.cos((i*np.pi/180)*(360/num_tracers)),0.1*np.sin((i*np.pi/180)*(360/num_tracers))]
        else:
            v0=[X_mat[i,4*j-1],Y_mat[i,4*j-1]]
        [x,y]=eval_xy(v0,t)
    
        #Now we will store all X's as a matrix and all Y's as a matrix. Each row corresponds to a tracer and each column corresponds on a time point.
        X_mat[i,4*j:4*j+4]=x
        Y_mat[i,4*j:4*j+4]=y
```

```python

plt.figure(figsize=(8,8))
for i in range(10):
    if(i==0):
        plt.scatter(X_mat[:,i*16],Y_mat[:,i*16],s=1)
    else: 
        plt.scatter(X_mat[:,i*16],Y_mat[:,i*16],s=1)
plt.scatter(X_mat[:,159],Y_mat[:,159],s=1)
'''

plt.scatter(X_mat[:,0],Y_mat[:,0],s=0.8,c="orange")
''''''plt.scatter(X_mat[:,20],Y_mat[:,20],s=0.8,c="green")
plt.scatter(X_mat[:,40],Y_mat[:,40],s=0.8,c="blue")
plt.scatter(X_mat[:,60],Y_mat[:,60],s=0.8,c="magenta")
plt.scatter(X_mat[:,80],Y_mat[:,80],s=0.8,c="gray")
plt.scatter(X_mat[:,99],Y_mat[:,99],s=0.8,c="cyan")'''
plt.legend(np.linspace(0,10,11),bbox_to_anchor=(1.05, 1))
plt.xlabel("$x_1$",size=16)
plt.ylabel("$x_2$",size=16)
plt.title("Q4(b) variation of the patch with time using 2500 tracer particles", size=16)
#plt.savefig("q4_b.png")
#plt.xlim((-3,3))
#plt.ylim((-3,3))
```
